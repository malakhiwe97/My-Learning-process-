```python
#Turning Questions into Analysis Plan

#Objective : identify factors that influenced passenger survival on the Titanic
#Approach :  Analyze survival rates across gender, age,family, class
```

```python
df['survived'].value_counts(normalize=True)
```

```python
#this tells us that the disaster was deadly
```

```python
df.groupby('sex')['survived'].mean()
```

```python
#Gender strongly influenced the survival
```

```python
df.groupby('pclass')['survived'].mean()
```

```python
#Socio-economic status played a role here
```

```python
df.groupby(['sex','pclass'])['survived'].mean()
```

```python
#Most females on class 1 survived while fewer nales survived on class 3
```

```python
df['age_group'] = pd.cut(df['age'], bins=[0,12,18,60,100], labels=['child','teenager','adult','senior'])
df.groupby('age_group')['survived'].mean()
```

```python
#Priority might have been given to a younger passenger
```

```python
df['family_size'] = df['sibsp'] + df['parch'] + 1
df.groupby('family_size')['survived'].mean()
```

```python
#Small families survived more tha solo or larger familes
```

```python
 df['is_alone'] = df['family_size'].apply(lambda x: 1 if x == 1 else 0)
 df.groupby('is_alone')['survived'].mean()
```

```python
df.groupby('survived')['fare'].mean()
```

```python
df.groupby('embarked')['survived'].mean()
```

```python
#Summary
#Survival on the Titanic was strongly influenced by gender, passenger class, age, family size, and ticket fare. Females, First class passengers and younger individuals, and those travelling in small families had significantly higher survival rates. These findings suggest that social economic status played a major role in determing survival outcomes
```
